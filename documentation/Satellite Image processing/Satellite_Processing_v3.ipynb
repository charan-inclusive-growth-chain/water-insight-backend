```python
# @title data functions
def ndci(image):
  return image.normalizedDifference(['B5', 'B4']).rename('NDCI')

def ndwi(image):
  return image.normalizedDifference(['B3','B5']).rename('NDWI')

def ndti(image):
  return image.normalizedDifference(['B4','B3']).rename('NDTI')

def ndsi(image):
  return image.normalizedDifference(['B11','B12']).rename('NDSI')

def pH(image):
  B1 = image.select('B1')
  B8 = image.select('B8')
  return B1.divide(B8).multiply(-0.827).add(8.339).rename('pH')

def dissolvedOxygen(image):
  B8 = image.select('B8')
  B9 = image.select('B9')
  B11 = image.select('B11')
  return B8.multiply(-0.0167).add(B9.multiply(0.0067)).add(B11.multiply(0.0083)).add(9.577).rename('DO')

def landTemp(image):
  B10 = image.select('ST_B10')
  return B10.multiply(0.00341802).add(149).subtract(273.15).rename('LandTemp')

def suspendedMatter(image):
  O8 = image.select('Oa08_radiance')
  O6 = image.select('Oa06_radiance')
  return O8.divide(O6).rename('SuspendedMatter')

def disOrgMatter(image):
  O8 = image.select('Oa08_radiance')
  O4 = image.select('Oa04_radiance')
  return O8.divide(O4).rename('DOM')
```

```python
# @title get data function
def getVals(start_date, end_date,aoi):

  s2_collection = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")\
                  .filterDate(start_date,end_date)\
                  .filterBounds(aoi)\
                  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE',75))

  s3_collection = ee.ImageCollection("COPERNICUS/S3/OLCI")\
                  .filterDate(start_date,end_date)\
                  .filterBounds(aoi)\
                  .sort('CLOUD_COVER')\
                  .limit(80)

  landsat_collection = ee.ImageCollection("LANDSAT/LC08/C02/T1_L2")\
                  .filterDate(start_date,end_date)\
                  .filterBounds(aoi)\
                  .sort('CLOUD_COVERAGE_ASSESSMENT')\
                  .limit(80)

  ndci_collection = s2_collection.map(lambda image: ndci(image.clip(aoi))).mean()
  ndwi_collection = s2_collection.map(lambda image: ndwi(image.clip(aoi))).mean()
  ndti_collection = s2_collection.map(lambda image: ndti(image.clip(aoi))).mean()
  ndsi_collection = s2_collection.map(lambda image: ndsi(image.clip(aoi))).mean()
  ph_collection = s2_collection.map(lambda image: pH(image.clip(aoi))).mean()
  do_collection = s2_collection.map(lambda image: dissolvedOxygen(image.clip(aoi))).mean()
  lt_collection = landsat_collection.map(lambda image: landTemp(image.clip(aoi))).mean()
  sm_collection = s3_collection.map(lambda image: suspendedMatter(image.clip(aoi))).mean()
  dom_collection = s3_collection.map(lambda image: disOrgMatter(image.clip(aoi))).mean()

  ndci2 = ndci_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )

  ndti2 = ndti_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )
  ndwi2 = ndwi_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )
  ndsi2 = ndsi_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )
  ph2 = ph_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )
  do2 = do_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )
  lt2 = lt_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )
  sm2 = sm_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )
  dom2 = dom_collection.reduceRegion(
      reducer=ee.Reducer.mean(),
      geometry=aoi,
      scale=100,
      maxPixels=1e9
  )

  ndci_val = 'Not Available' if not bool(ndci2.getInfo()) else ndci2.getInfo()['NDCI']
  ndti_val = 'Not Available' if not bool(ndti2.getInfo()) else ndti2.getInfo()['NDTI']
  ndsi_val = 'Not Available' if not bool(ndsi2.getInfo()) else ndsi2.getInfo()['NDSI']
  ndwi_val = 'Not Available' if not bool(ndwi2.getInfo()) else ndwi2.getInfo()['NDWI']
  ph_val = 'Not Available' if not bool(ph2.getInfo()) else ph2.getInfo()['pH']
  do_val = 'Not Available' if not bool(do2.getInfo()) else do2.getInfo()['DO']
  lt_val = 'Not Available' if not bool(lt2.getInfo()) else lt2.getInfo()['LandTemp']
  sm_val = 'Not Available' if not bool(sm2.getInfo()) else sm2.getInfo()['SuspendedMatter']
  dom_val = 'Not Available' if not bool(dom2.getInfo()) else dom2.getInfo()['DOM']

  #global df2
  #df2 = df2.append({'Start Date':start_date, 'End Date':end_date, 'NDCI':ndci_val, 'NDTI':ndti_val, 'NDSI':ndsi_val, \
  #            'NDWI':ndwi_val,'pH': ph_val, 'DO':do_val, 'LandTemp':lt_val,\
  #           'SuspendedMatter':sm_val, 'DOM':dom_val}, ignore_index=True)
  return ndci_val, ndti_val, ndsi_val, ndwi_val, ph_val, do_val, lt_val, sm_val, dom_val
```

```python
def getData(aoi):
  end_date = date.today()
  start_date = end_date+timedelta(days=-5)
  end_date = str(end_date.strftime('%Y-%m-%d'))
  start_date = str(start_date.strftime('%Y-%m-%d'))
  ndci_val, ndti_val, ndsi_val, ndwi_val, ph_val, do_val, lt_val, sm_val, dom_val = getVals(start_date, end_date, aoi)
  return end_date, start_date, ndci_val, ndti_val, ndsi_val, ndwi_val, ph_val, do_val, lt_val, sm_val, dom_val
```

```python
def getCustomData(start_date, end_date, aoi):
  ndci_val, ndti_val, ndsi_val, ndwi_val, ph_val, do_val, lt_val, sm_val, dom_val = getVals(start_date, end_date, aoi)
  return end_date, start_date, ndci_val, ndti_val, ndsi_val, ndwi_val, ph_val, do_val, lt_val, sm_val, dom_val
```

```python
def getOneYearData(aoi):
    date_itr_end = date.today()
    date_itr_start = date_itr_end+timedelta(days=-370)
    date_itr_end = str(date_itr_end.strftime('%Y-%m-%d'))
    date_itr_start = str(date_itr_start.strftime('%Y-%m-%d'))
    date_range = pd.date_range(date_itr_start, date_itr_end, freq='1D')
    df = pd.DataFrame(columns = ['lake', 'End Date','Start Date', 'NDCI', 'NDTI', 'NDSI', 'NDWI', 'pH', 'DO', 'LandTemp', 'SuspendedMatter', 'DOM'])

    aoi_name = ''
    if (aoi == aoi_osman):
      aoi_name = 'Osman Sagar'
    else:
      aoi_name = 'Durgam Cheruvu'

    for i in range(5,len(date_range)):
        ndci_val, ndti_val, ndsi_val, ndwi_val, ph_val, do_val, lt_val, sm_val, dom_val = getVals(date_range[i-5], date_range[i], aoi)
        df = df.append({'lake':aoi_name, 'Start Date':date_range[i-5], 'End Date':date_range[i], 'NDCI':ndci_val, 'NDTI':ndti_val, 'NDSI':ndsi_val, \
              'NDWI':ndwi_val,'pH': ph_val, 'DO':do_val, 'LandTemp':lt_val,\
             'SuspendedMatter':sm_val, 'DOM':dom_val, }, ignore_index=True)

    return df
```
